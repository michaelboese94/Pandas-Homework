```python
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
```

```python
purchase_df.head()
```

```python
#Player count 
total_players = purchase_df["SN"].unique()
tot_players = len(total_players)
total_players_dict = {"Total Players":[total_players]}
total_players_df = pd.DataFrame.from_dict(tot_players_dict)
total_players_df
```

```python
# Purchasing Analysis(total)
#unique values 
unique_purchase = purchase_df["Item ID"].unique()
total_unique = len(unique_purchase)

#Average Purchase Price
avg_price = purchase_df["Price"].mean()

#Total number of purchases
total_num_purchase = len(purchase_df["Purchase ID"])

#Total Revenue 
total_revanue = purchase_df["Price"].sum()

#Creating the output
purchase_analysis_dict = {"Unique Values" : [total_unique],
                          "Average Price" : [avg_price], 
                          "Total Number of Purchases":total_num_purchase], 
                          "Total Revanue" : [total_revanue]}
purchase_analysis_df = pd.DataFrame.from_dict(purchase_analysis_dict)
purchase_analysis_round = purchase_analysis_df.round(decimals = 2)
purchase_analysis_round
```

```python
#Gender Demographics 
drop_gender_df = pd.DataFrame(purchase_df.drop_duplicates(["SN"], keep = "first", inplace = False)) 
total = len(drop_gender_df["Gender"])

# Male, Female and other  total values 
male = drop_gender_df.groupby(["Gender"]).get_group(('Male'))
total_male = len(male)
female = drop_gender_df.groupby(["Gender"]).get_group(('Female'))
total_female = len(female)
other = drop_gender_df.groupby(["Gender"]).get_group(('Other / Non-Disclosed'))
total_other = len(other)

#Male, Female and other percent values
percent_male = (total_male / total) *100
percent_female = (total_female / total) *100
percent_other = (total_other / total) *100

#Creating the output
gender_dict = {"Gender" : ["Male", "Female", "Other / Non-Disclosed"], 
               "Total Count" : [total_male, total_female, total_other],
                "Percentage of Players" : [percent_male, percent_female, percent_other]}
gender_df = pd.DataFrame.from_dict(gender_dict)
gender_round = gender_df.round(decimals = 2)
gender_round

```

```python
#Purchase Analysis(Gender)
#Calculating the purchasing totals 
male1 = purchase_df.groupby(["Gender"]).get_group(('Male'))
total_male1 = len(male1)
female1 = purchase_df.groupby(["Gender"]).get_group(('Female'))
total_female1 = len(female1)
other1 = purchase_df.groupby(["Gender"]).get_group(('Other / Non-Disclosed'))
total_other1 = len(other1)

#Calculating the averages by gender
avg_male = male1["Price"].mean()
avg_female = female1["Price"].mean()
avg_other = other1["Price"].mean()

#calculating the average total purchases 
sum_male = male1["Price"].sum()
sum_female = female1["Price"].sum()
sum_other = other1["Price"].sum()

avg_total_male = sum_male / total_male
avg_total_female = sum_female / total_female
avg_total_other = sum_other / total_other

#creating the data frame for average total purchases
per_person_dict = {"Genders" : ['Male' , 'Female' , 'Other / Non-Disclosed'],
                  "Average Purchase Price" : [avg_male , avg_female , avg_other],
                   "Total Purchase Value" : [sum_male, sum_female, sum_other],
                  "Average Total Purchase per Person" : [avg_total_male , avg_total_female , avg_total_other]}
per_person_df = pd.DataFrame.from_dict(per_person_dict)
per_person_round = per_person_df.round(decimals = 2)
per_person_round
```
